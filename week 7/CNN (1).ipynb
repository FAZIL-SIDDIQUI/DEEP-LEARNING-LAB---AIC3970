```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import skimage.transform as transform

# Activation Functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

activation_functions = {
    'relu': relu,
    'sigmoid': sigmoid,
    'tanh': tanh
}

# 2D Convolution
def conv2d(image, kernel, stride):
    image_h, image_w = image.shape
    kernel_h, kernel_w = kernel.shape
    out_h = (image_h - kernel_h) // stride + 1
    out_w = (image_w - kernel_w) // stride + 1
    output = np.zeros((out_h, out_w))
    
    for i in range(out_h):
        for j in range(out_w):
            region = image[i * stride : i * stride + kernel_h, j * stride : j * stride + kernel_w]
            output[i, j] = np.sum(region * kernel)
    return output

# Max Pooling
def maxpool2d(feature_map, pool_size, stride):
    fm_h, fm_w = feature_map.shape
    out_h = (fm_h - pool_size) // stride + 1
    out_w = (fm_w - pool_size) // stride + 1
    pooled = np.zeros((out_h, out_w))
    
    for i in range(out_h):
        for j in range(out_w):
            region = feature_map[i * stride : i * stride + pool_size, j * stride : j * stride + pool_size]
            pooled[i, j] = np.max(region)
    return pooled

# Dense Layer
def dense_layer(inputs, weights, bias):
    return np.dot(inputs, weights) + bias

# Main CNN Process
def main():
    num_filters = int(input("Enter the number of filters for each convolution: "))
    conv_stride = int(input("Enter the stride value for the convolution: "))
    act_choice = input("Enter activation function (relu, sigmoid, tanh): ").lower()
    
    activation = activation_functions.get(act_choice, relu)
    
    digits = load_digits()
    image = digits.images[0]
    image = transform.resize(image, (10, 10), anti_aliasing=True)
    
    plt.imshow(image, cmap='gray')
    plt.title("Preprocessed 10x10 Image")
    plt.axis('off')
    plt.show()
    
    print("\nPreprocessed Input Image (10x10):\n", image)
    
    kernel_sizes = [(2, 2), (3, 3), (5, 5)]
    conv_results = []

    for k_index, k_size in enumerate(kernel_sizes):
        print(f"\nApplying convolution with kernel size {k_size}:")
        filters_outputs = []
        for f in range(num_filters):
            kernel = np.random.rand(k_size[0], k_size[1])
            conv_out = conv2d(image, kernel, conv_stride)
            conv_act = activation(conv_out)
            filters_outputs.append(conv_act)
            
            print(f"\nFilter {f+1} Output before activation (shape={conv_out.shape}):")
            print(conv_out)
            print(f"\nFilter {f+1} Output after activation:")
            print(conv_act)
            
            plt.imshow(conv_act, cmap='viridis')
            plt.title(f"Conv Act - Kernel {k_size}, Filter {f+1}")
            plt.colorbar()
            plt.show()
        conv_results.append(filters_outputs)

    # Max Pooling Step
    pool_size = 2
    pool_stride = 2
    pooled_feature_maps = []

    for group_index, group in enumerate(conv_results):
        pooled_group = []
        for filter_index, feature_map in enumerate(group):
            pooled = maxpool2d(feature_map, pool_size, pool_stride)
            pooled_group.append(pooled)

            print(f"\nPooled Feature Map (Kernel {kernel_sizes[group_index]}, Filter {filter_index+1}):")
            print(pooled)

            plt.imshow(pooled, cmap='inferno')
            plt.title(f"Max Pool - Kernel {kernel_sizes[group_index]}, Filter {filter_index+1}")
            plt.colorbar()
            plt.show()
        pooled_feature_maps.append(pooled_group)

    # Flattening Step
    flattened = []
    for group in pooled_feature_maps:
        for feature_map in group:
            flattened.extend(feature_map.flatten())
    flattened = np.array(flattened)
    print("\nFlattened feature vector shape:", flattened.shape)

    # Dense Layer (10 units)
    dense_units = 10
    flattened_dim = flattened.shape[0]
    dense_weights = np.random.randn(flattened_dim, dense_units)
    dense_bias = np.random.randn(dense_units)

    dense_out = dense_layer(flattened, dense_weights, dense_bias)
    dense_act = activation(dense_out)

    print("\nDense Layer Output (10 units):\n", dense_act)

    # Output Layer (1 unit)
    out_weights = np.random.randn(dense_units, 1)
    out_bias = np.random.randn(1)
    final_output = dense_layer(dense_act, out_weights, out_bias)

    print("\nFinal Output (1 unit):", final_output)

if __name__ == '__main__':
    main()
```
